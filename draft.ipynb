```python
import nltk
import pymorphy2
import string
morph = pymorphy2.MorphAnalyzer()
```

```python
sting = string.punctuation + '1234567890qwertyuiopasdfghjklzxcvbnm'
```

```python
filepath = r"C:\Users\Пользователь\универ\infos\friends-data\Friends - season 7\Friends - 7x01 - The One With Monica's Thunder.ru.txt"
words = {}
with open(filepath, 'r', encoding = 'UTF-8') as f:   
    text = f.read().lower()
text = text[1:]
for s in sting:
    text = text.replace(s,'')
pretext = nltk.word_tokenize(text) 
for i, p in enumerate(pretext):   
    pretext[i] = morph.parse(p)[0][2] 
    p = pretext[i]
    if p not in words:
        words[p] = {}
        words[p][filepath] = 1
    else:
        words[p][filepath] += 1
line = ' '.join(pretext)
print(words)
```
